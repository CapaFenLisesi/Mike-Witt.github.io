<a name=bcalc></a>
<h2><a href="#top">^</a> Basic Calculations</h2>
<p>
    Note that: Superscripts denote the reference frame and subscripts the event. Also, $x$ is position and $t$ is time. So, for example, $x^{bf}_{am}$ is the position of Alice's measurement viewed by Bob.
</p>
    

```python
# 1 light second = speed of light * 1 second
LS = c*1
bf_x_bm = 0
bf_t_bm = 0
bf_x_am = 10*LS
bf_t_am = 1

# When I print the distances below, I divide by LS so that the results
# will be diplayed in light seconds.

# Ted is going -.6c in Bob's frame. So, in Ted's frame
tf_x_bm = ltx(bf_x_bm, bf_t_bm, -.6*c)
tf_t_bm = ltt(bf_x_bm, bf_t_bm, -.6*c)
print('Ted sees Bob\'s measurement at x=%s, t=%s'
      %(tf_x_bm/LS, tf_t_bm))
tf_x_am = ltx(bf_x_am, bf_t_am, -.6*c)
tf_t_am = ltt(bf_x_am, bf_t_am, -.6*c)
print('Ted sees Alice\'s measurement at x=%.2f, t=%.2f'
      %(tf_x_am/LS, tf_t_am))
cf_x_am = ltx(bf_x_am, bf_t_am, .6*c)
cf_t_am = ltt(bf_x_am, bf_t_am, .6*c)
print('Carol sees Alice\'s measurement at x=%.2f, t=%.2f'
      %(cf_x_am/LS, cf_t_am))
```

<p>
    Now, let's say that Alice is moving .6c in Bob's frame. At t=0 (for both of them) Alice passes Bob
    and (without slowing down :-) picks up her part of the entangled pair. Then, in Bob's frame, (1)
    Bob does his measurement at t=10 when he sees Alics at 10 LS. (2) Bob receives her answer at
    22 seconds, seeing that she sent it at 11 sec.
</p>

```python
# 1 light second = speed of light * 1 second
LS = c*1
bf_x_bm = 0
bf_t_bm = 10
bf_x_am = 11*LS
bf_t_am = 11

# When I print the distances below, I divide by LS so that the results
# will be diplayed in light seconds.

# Alice is going .6c in Bob's frame. So, in Alice's frame
af_x_bm = ltx(bf_x_bm, bf_t_bm, .6*c)
af_t_bm = ltt(bf_x_bm, bf_t_bm, .6*c)
print('Alice sees Bob\'s measurement at x=%s, t=%s'
      %(af_x_bm/LS, af_t_bm))
af_x_am = ltx(bf_x_am, bf_t_am, .6*c)
af_t_am = ltt(bf_x_am, bf_t_am, .6*c)
print('Alice sees Alice\'s measurement at x=%.2f, t=%.2f'
      %(af_x_am/LS, af_t_am))
```

<h2><a href="#top">^</a> SpaceTime Diagrams</h2>

```python
# time axis in the Bob's frame
plot_t0 = -10
nt = 11. # Number of points to plot
plot_t1 = 10 # end time in seconds
t_axis = linspace(plot_t0, plot_t1, nt)
v = .6*c
xlow = -3; xhigh = 22
tlow = plot_t0; thigh = plot_t1+2

bnds = (-3, 22, plot_t0/c, plot_t1/c)

# The paths. Everything is in light seconds
Ox = 0; # The origin. This is where Bob does his measurement
bob_path = []; alice_path = []; 
ted_path = []; carol_path = []
for t in t_axis:
    bob_path += [ Ox, ] # was track_l
    alice_path += [ Ox+10, ] # was track_r
    ted_path += [ Ox, ] # Ted's path in Ted's frame
    carol_path += [ Ox, ] # Carol's path in Carol's frame

# Make world lines in their respective frames
Bob_wl = world_line('Bob', bob_path, t_axis, 'black') # was track_l
Alice_wl = world_line('Alice', alice_path, t_axis, 'black') # was track_r
Ted_wl = world_line('Ted', ted_path, t_axis, 'green')
Carol_wl = world_line('Carol', carol_path, t_axis, 'orange')

# Then translate them for Bob's frame
bob_wls = []
bob_wls += [ Bob_wl, ]
bob_wls += [ Alice_wl, ]
bob_wls += [ Ted_wl.ltrans(-v), ]
bob_wls += [ Carol_wl.ltrans(v), ]

# Events in the Bob's frame frame:
#  E_BM was E2
#  E_AM was E1
E_BM= event('Bob Meas', 0, 0, 'red') # Bob's Measurement
E_AM= event('Alice Meas', 10, 1/c, 'blue') # Alice's Measurement (in LS)
bob_evs = [E_BM, E_AM]

# Create and display Bob's frame
bob_frame = frame('Bob\'s Frame', bob_wls, bob_evs)
# IF you want to display a "grid" on the frame
if True:
    gbnds = [bnds[0]*4, bnds[1]*4, bnds[2]*4, bnds[3]*4]
    grids = []
    xs = 1.0; ts = 1.0
    bob_grid = grid(bounds=gbnds,
        x_spacing=xs, t_spacing=ts/c, color='black', thickness=.2)
    grids += [ bob_grid, ]
    ted_grid = grid(bounds=gbnds,
        x_spacing=xs, t_spacing=ts/c, color='green', thickness=.2)
    #grids += [ ted_grid.ltrans(-v), ]
    carol_grid = grid(bounds=gbnds,
        x_spacing=xs, t_spacing=ts/c, color='orange', thickness=.4)
    #grids +=[ carol_grid.ltrans(v), ]
    bob_frame.grids = grids
pyplot = bob_frame.display(figsize=(8,8), bounds=bnds, scale='meters')
foo=pyplot.xlim(bnds[0], bnds[1]) # Shouldn't need this!
foo=pyplot.ylim(bnds[2], bnds[3]) # Shouldn't need this!
```

```python
# World lines and events for the Ted's frame
ted_wls = []
ted_wls += [ Bob_wl.ltrans(-v), ]
ted_wls += [ Alice_wl.ltrans(-v), ]
ted_wls += [ Ted_wl, ]
# Carol's velocity seen by Ted ...
# (formula here)
#ted_wls += [ Carol_wl, vc]
ted_evs = []
ted_evs += [ E_AM.ltrans(-v), ]
ted_evs += [ E_BM.ltrans(-v), ]

# Create and display Ted's frame
ted_frame = frame('Ted\'s Frame', ted_wls, ted_evs)
# IF you want to display a "grid" on the frame
if True:
    gbnds = [bnds[0]*4, bnds[1]*4, bnds[2]*4, bnds[3]*4]
    grids = []
    xs = 1.0; ts = 1.0
    bob_grid = grid(bounds=gbnds,
        x_spacing=xs, t_spacing=ts/c, color='black', thickness=.3)
    #grids += [ bob_grid.ltrans(v), ]
    ted_grid = grid(bounds=gbnds,
        x_spacing=xs, t_spacing=ts/c, color='green', thickness=.3)
    grids += [ ted_grid, ]
    ted_frame.grids = grids
pyplot = ted_frame.display(figsize=(8,8), bounds=bnds, scale='meters')
foo=pyplot.xlim(bnds[0], bnds[1]) # Shouldn't need this!
foo=pyplot.ylim(bnds[2], bnds[3]) # Shouldn't need this!
```
