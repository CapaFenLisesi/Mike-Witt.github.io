<a name=basics></a>
<h2><a href="#top">^</a> Basics</h2>

The spacetime diagrams that we'll draw here have one dimension of space
and one dimension of time. This means that they can only model motion along
the x axis, a single line.
<p>
A location in spacetime is called an "event." In 4D spacetime, an event has
four coordinates: x, y, z, and t. But since our diagrams only have one
spatial dimension, our events just have the x and t coordinates.
<p>
In choosing the "units" for the space and time axes, we have to decide
whether we want a picture that shows "human scale" events (in which case
we might have meters along the x axis and seconds along the time axis)
or events where the objects involved are travelling at some significant
fraction of the speed of light (in which case we want both space and time
to have the same measure -- meters for example).
<p>
Normally, we will take the 2nd option, using either seconds or meters for
<i>both</i> the space and time axes. This has the (desirable) side-effect
of "scaling" things so that the speed of light equals 1.
<p>
In this, and the following sections, I'll put comments in the code to
explain specific details. So, read through the comments to get an idea
how the library routines are being used to draw the diagrams.

<h3>Drawing a reference frame</h3>
<p>

```python
# Create an empty frame
a_frame = sr.frame('A Frame')

# Get the speed-of-light constant from srlib
c = sr.c

# Here we set the boundary of the plot. There are some things to be
# aware of here:
#   (1) All the numbers should be entered with a decimal point
#   (2) In this example we are doing everything in meters
#   (3) In order for "meters" to work as time, we need to divide both
#       of the time values by the speed of light
low_x = 0.0; high_x = 10.0
low_t = 0.0; high_t = 10.0
bounds = (low_x, high_x, low_t/c, high_t/c)

# Now we display the frame
#    scale: tells whether we are using meters, seconds, etc
#    bounds: is the boundary definition
#    figsize: tells how big to draw the plot
#    nticks: is the number of labels along the time axis
a_frame.display(scale='meters', bounds=bounds, figsize=[6,6], nticks=6);
```

<h3>Adding some events</h3>
<p>

```python
# Now we add some events to the diagram. Just like when
# specifying the boundary, if we are working in meters,
# we need to divide all the time coordinates by c.
#
# The syntax is:
#
#  event_variable = event(name, space_coordinate, time_coordinate, color)

# The next three lines define three events
E1 = sr.event('E1', 2., 2./c, 'red')
E2 = sr.event('E2', 3., 5./c, 'blue')
E3 = sr.event('E3', 8., 6./c, 'green')

# This line adds the events to the frame, using the same bounds as before
a_frame.events = [E1, E2, E3]

# And we display the frame
a_frame.display(scale='meters', bounds=bounds, figsize=[6,6], nticks=6);
```

<a name=worldlines></a>
<h2><a href="#top">^</a> World Lines</h2>
<p>
A <i>world line</i> is simply a path through spacetime. I didn't make
up the name :-)
<p>
You could draw your world line for a day, starting with the event of you
waking up in the morning and ending with the event of you going to sleep
the following night. This would result in a jagged (non inertial) path
through space and time, charting your movements during the day.
<p>
Typically we draw the world line of a person or
object from the beginning to the end of some experiment.
<p>
Below, I'll draw Alice's frame of reference, with her world line in
it. From Alice's perspective, she simply stands at the origin of the
world and doesn't move (though her world line does extend forward
in time).

```python
# Make a frame
alices_frame = sr.frame("Alice's Frame")

#
# Making a world line
#
# First we'll make a time axis. This is just a list of numbers
# specifying every time we care about. As usual, we'll divide
# t by the speed of light, to get it in meters.
#
t0 = 0.0    # start the time at zero
t1 = 10.0/c # End the time at 10 (meters)
nt = 11     # number of time points to use
# The next line creates an array of times (the "t axis"). We use
# a function called "linspace" from the standard numpy library.
t_axis = np.linspace(t0, t1, nt)
# Now, we create an array of positions along the x axis for Alice.
# The next line just makes an empty array:
alice_positions = []
# And the following "for loop" adds a position for each time  in
# the time array. Since Alice is standing still at the origins,
# so all the x positions are zero.
for t in t_axis:
    alice_positions += [ 0.0, ]
# Here are the arrays:
print t_axis*c
print alice_positions
```

```python
# The "world_line" function from srlib creates the actual world line.
alice_wl = sr.world_line('Alice', alice_positions, t_axis, 'green')
# Here we put it in Alice's frame
wlines = [alice_wl]
alices_frame.wlines = wlines
# And display the frame. We make the bounds large enough to show
# the world line.
bounds = [-5, 5, 0, 10.0/c]
alices_frame.display(scale='meters', bounds=bounds, figsize=[6,6]);
```

```python
# Let's say that in Bob's frame of reference, he is also standing
# still at x=0. We'll create his frame and world line, just like 
# we did for Alice.

bobs_frame = sr.frame("Bob's Frame")
bob_positions = []
for t in t_axis:
    bob_positions += [0,]
bob_wl = sr.world_line('Bob', bob_positions, t_axis, 'red')
bobs_frame.wlines = [bob_wl]
bobs_frame.display(scale='meters', bounds=bounds, figsize=[6,6]);
```

```python
# Suppose Bob is moving at half the speed of light, to the right,
# relative to Alice:
v = c/2
# We can transform Bob's world line into Alices perspective like this:
bob_in_alices_frame = bob_wl.ltrans(-v)
# And show both world lines in Alice's frame:
alices_frame.wlines = [alice_wl, bob_in_alices_frame]
alices_frame.display(scale='meters', bounds=bounds, figsize=[6,6]);
```

```python
# And here is Alices in Bob's Frame:
bobs_frame.wlines = [bob_wl, alice_wl.ltrans(v)]
bobs_frame.display(scale='meters', bounds=bounds, figsize=[6,6]);
```

<a name=grids></a>
<h2><a href="#top">^</a> Grids</h2>
<p>
The last thing I'll do in this notebooks is to demonstrate making a
grid of time and space lines. I won't go into this in detail. We can
talk about it more in the meeting. But, basically, here is a grid of
Alice's time and space coordinates. Each horizontal line represents
a given time for Alice, and each vertical line represents a given
spatial location.

```python
#bndsX = []
#for b in bounds: bndsX += [b*10]
bndsX = [-20, 10, -20/c, 20/c]
alice_grid = sr.grid(
    bounds=bndsX, x_spacing=1, t_spacing=1/c, 
    color='green', thickness=.5)
alices_frame.grids = [alice_grid]
alices_frame.display(scale='meters',bounds=bounds,figsize=[6,6],nticks=6);
```

I can also transform Bob's time/space grid so that I can show it in Alice's frame of reference. This demonstrates the "relativity of simultaneity" by showing how the line that represents a specific instant in time for Bob is very different than for Alice.

```python
bob_grid = sr.grid(
    bounds=bndsX, x_spacing=1, t_spacing=1/c, 
    color='red', thickness=.5)
alices_frame.grids = [alice_grid, bob_grid.ltrans(-v)]
alices_frame.display(scale='meters',bounds=bounds,figsize=[6,6],nticks=6);
```
