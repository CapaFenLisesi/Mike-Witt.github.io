<h2>Two functions, f and g</h2>
<p class=in>
    $f(x) = (x-5)^2$
    <br><br>
    $g(x) = x-3$
</p>

```python
# Define the two functions
def f(x):
    return( (x-5)**2 )
def g(x):
    return( x-3 )

# Think of this, for now, as "creating the x axis"
x_axis = linspace(2, 8, 100)

# Then we plot the two functions on the same axis
plot(x_axis, f(x_axis), color='green')
plot(x_axis, g(x_axis), color='red')
ylimits = ylim(0, 7) # This line just make the plot look better
```

<p>
    Notice how there are two points on the graph where $f(x)=g(x)$. We can find those points "analytically"
    in the following way. First "multiply out" the function $f$:
</p>
<p class=in>
    $(x-5)^2 = x^2 - 10x + 25$
</p>
Then set the two functions equal to one another:
<p class=in>
    $x^2 - 10x + 25 = x-3$
</p>
<p>
    Add $3$ to both sides:
</p>
 <p class=in>
    $x^2 - 10x + 28 = x$
</p> 
<p>
    Subtract $x$ from both sides:
</p>
 <p class=in>
    $x^2 - 11x + 28 = 0$
</p> 
<p>
    This expression can be factored:
</p>
 <p class=in>
    $x^2 - 11x + 28 = (x-4)(x-7)= 0$
</p>
<p>
    And the result will be zero when either $x=4$ or $x=7$.
</p>
<p>
    The answers can be "checked" by plugging the two proposed
    answers into both the functions:
</p>
<p class=in>
    $f(4) = (4-5)^2 = (-1)^2 = 1$
    <br><br>
    $g(4) = 4-3 = 1$
    <br><br>
    $f(7) = (7-5)^2 = (2)^2 = 4$
    <br><br>
    $g(7) = 7-3 = 4$
</p>
<p>
We can also plot the two proposed answers as points on the graph like this:
</p>

```python
# Draw the functions just the same as before
plot(x_axis, f(x_axis), color='green')
plot(x_axis, g(x_axis), color='red')
# Points can be plotted as follows. The first list of numbers
# contains the x coordinates and the second list contains the
# y coordinates. 'o' basically says to plot individual points.
plot([4, 7], [1, 4], 'o', color='black')
ylimits = ylim(0, 7) # Again, limit the plot to this y range.
```
