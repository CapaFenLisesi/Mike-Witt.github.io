<h2>Photon bounces off a mirror</h2>
</center>

This is a "space-time diagram" of a scenario where Alice
reflects a photon off a mirror and Bob watches, but it is <i>not</i>
the same setup that we used to
derive the formula for time dilation.


<a name=setup></a>
<h2>Setup</h2>
<p>
I've got a whole bunch of code in the setup cell, so that I don't
have to interrupt the flow of the text and diagrams. For the time
being just ignore the code here.

```python
%autosave 0
%matplotlib inline
import numpy as np
from srlib import *

# time axis in the Alice's frame
t0 = 0.
nt = 501. # Number of points to plot
t1 = 10./c # End time in metres
t_axis = np.linspace(t0-t1*.5, t1+t1*.5, nt)

# The paths in Alice's frame at time t_a
alice = []; mirror = []; photon = []
for t in t_axis:
    alice += [1.0,] # Alice's path
    mirror += [6.0,] # Path of the mirror
    # Path of the photon. The "1+" is so that the photon's path starts
    # at Alice's position instead of the origin.
    if t<0 or t>t1: photon += [ np.nan, ] # "not a number." Don't display.
    elif t<=t1/2: photon += [1 + c*t,]
    else: photon += [1 + c*t1/2 - c*(t-t1/2),]

# Make world lines for the paths
wlines = []
wlines += [ world_line('Alice', alice, t_axis, 'green'), ]
wlines += [ world_line('Mirror', mirror, t_axis, 'black'), ]
wlines += [ world_line('Photon', photon, t_axis, 'orange'), ]
```

<div style="Page-break-after: always;"></div></code>
<a name=photon></a>
<h2>Alice releases a photon</h2>
<p>
We have a space axis running to the right, and a time
axis running up. From Alice's perspective both she (the green
line) and the mirror (the black line) are standing still. So
neither line moves along the space axis. But, of course, both
Alice and the mirror move through time (straight up).

```python
alice_frame = frame('Alice\'s Frame', wlines, events=[], grids=[])
bnds = (.5, 6.5, -1./c, 11./c)
alice_frame.display(scale='metres', bounds=bnds, V=False);
```

<div style="Page-break-after: always;"></div></code>
<h2>In Bob's reference frame</h2>

As seen from Bob's perpective, both Alice and the mirror are moving
along the space axis as well as through time. So both of the world-lines
are slanted, rather than going straight up. Note that light itself will
always make the same 45$^\circ$ angle.
<p>
Notice how in Bob's frame the light path through spacetime is much
longer on the journey to the mirror than on the way back to Alice.
This is because this is
<i>not</i> the same experiment we did on the
board. That experiment actually had 2 dimensions. (1) The direction
the photon travelled to the mirror and back, and (2) The direction of
Alices travel relative to Bob. In <i>this</i> picture, Alice is releasing
the photon toward a mirror that is in the same direction as her travel
relative to Bob.

```python
bob_frame = frame('Bob\'s Frame', alice_frame.wlines, v=-.5*c)
pyplot = bob_frame.display(scale='metres', nticks=8, figsize=[7,7])
pyplot.ylim(-1/c, 18/c);
```
